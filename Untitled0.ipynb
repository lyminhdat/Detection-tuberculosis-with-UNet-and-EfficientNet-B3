```python
# Cài đặt các thư viện cần thiết
import os
import numpy as np
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, Concatenate

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping

from tensorflow.keras.applications.mobilenet import preprocess_input
# Kiểm tra GPU
print("GPU detected: ", tf.config.list_physical_devices('GPU'))
```

```python
import pandas as pd # data processing
from PIL import Image
from tensorflow import keras
from scipy import io
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
```

```python
# Hàm tải ảnh
def load_images(path):
    import cv2
    images = []
    for file_name in sorted(os.listdir(path)):
        img = cv2.imread(os.path.join(path, file_name), cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (256, 256))  # Resize về 256x256
        images.append(img)
    return np.array(images)
```

```python
# Đường dẫn dữ liệu phân đoạn
images_path = '/content/drive/MyDrive/Test_KLTN/Lung Segmentation/CXR_png'
masks_path = '/content/drive/MyDrive/Test_KLTN/Lung Segmentation/masks'
```

```python
# Tải dữ liệu phân đoạn
x_data = load_images(images_path)
y_data = load_images(masks_path)

x_data = x_data / 255.0  # Chuẩn hóa về [0, 1]
y_data = y_data / 255.0  # Chuẩn hóa về [0, 1]
```

**Data Augmentation**

```python
# Tăng cường dữ liệu
x_flip = np.flip(x_data , axis = 2)
y_flip = np.flip(y_data, axis = 2)

x_data = np.append(x_data, x_flip, axis = 0)
y_data = np.append(y_data, y_flip, axis = 0)

print("X Shape: ", x_data.shape, "Y Shape: ", y_data.shape)
```

```python
# Chia tập dữ liệu phân đoạn
x_train, x_temp, y_train, y_temp = train_test_split(x_data, y_data, test_size=0.3, random_state=42)

x_val, x_test, y_val, y_test = train_test_split(x_temp, y_temp, test_size=0.33, random_state=42)  # 20% kiểm định, 10% kiểm tra
```

**Xây dựng mô hình U-Net**

```python
# Xây dựng mô hình Unet

inputs = layers.Input(shape=(256,256,1))

c0 = layers.Conv2D(64, activation='relu', kernel_size=(3,3), padding='same', name = 'C0')(inputs)
Drop = layers.Dropout(0.1)(c0)
c1 = layers.Conv2D(64, activation='relu', kernel_size=(3,3), padding='same', name = 'C1')(Drop)
c2 = layers.MaxPool2D(pool_size=(2, 2), name = 'MaxPool_1')(c1)

c3 = layers.Conv2D(128, activation='relu', kernel_size=(3,3), padding='same', name = 'C3')(c2)
Drop = layers.Dropout(0.1)(c3)
c4 = layers.Conv2D(128, activation='relu', kernel_size=(3,3), padding='same', name = 'C4')(Drop)
c5 = layers.MaxPool2D(pool_size=(2, 2), name = 'MaxPool_2')(c4)

c6 = layers.Conv2D(256, activation='relu', kernel_size=(3,3), padding='same', name = 'C6')(c5)
Drop = layers.Dropout(0.2)(c6)
c7 = layers.Conv2D(256, activation='relu', kernel_size=(3,3), padding='same', name = 'C7')(Drop)
c8 = layers.MaxPool2D(pool_size=(2, 2), name = 'MaxPool_3')(c7)

c9 = layers.Conv2D(512, activation='relu', kernel_size=(3,3),padding='same',  name = 'C9')(c8)
Drop = layers.Dropout(0.2)(c9)
c10 = layers.Conv2D(512, activation='relu', kernel_size=(3,3), padding='same', name = 'C10')(Drop)
c11 = layers.MaxPool2D(pool_size=(2, 2), name = 'MaxPool_4')(c10)

c12 = layers.Conv2D(1024, activation='relu', kernel_size=(3,3), padding='same', name = 'C12')(c11)
Drop = layers.Dropout(0.3)(c12)
c13 = layers.Conv2D(1024, activation='relu', kernel_size=(3,3), padding='same', name = 'C13')(Drop)

###
t01 = layers.Conv2DTranspose(512, kernel_size=(2,2), strides=(2,2), padding='same', name = 'Transpose_T01')(c13)
concat01 = layers.concatenate([t01, c10], axis=3, name= 'Concat_1')

c14 = layers.Conv2D(512, activation='relu', kernel_size=(3,3), padding='same', name = 'C14')(concat01)
Drop = layers.Dropout(0.3)(c14)
c15 = layers.Conv2D(512, activation='relu', kernel_size=(3,3), padding='same', name = 'C15')(Drop)

t02 = layers.Conv2DTranspose(256, kernel_size=(2,2), strides=(2,2), padding='same', name = 'Transpose_T02')(c15)
concat02 = layers.concatenate([t02, c7], axis=3, name= 'Concat_2')

c16 = layers.Conv2D(256, activation='relu', kernel_size=(3,3), padding='same', name = 'C16')(concat02)
Drop = layers.Dropout(0.2)(c16)
c17 = layers.Conv2D(256, activation='relu', kernel_size=(3,3), padding='same', name = 'C17')(Drop)

t03 = layers.Conv2DTranspose(128, kernel_size=(2,2), strides=(2, 2), name = 'Transpose_T03')(c17)
concat03 = layers.concatenate([t03, c4], axis=3, name= 'Concat_3')

c18 = layers.Conv2D(128, activation='relu', kernel_size=(3,3), padding='same', name = 'C18')(concat03)
Drop = layers.Dropout(0.2)(c18)
c19 = layers.Conv2D(128, activation='relu', kernel_size=(3,3), padding='same', name = 'C19')(Drop)

t04 = layers.Conv2DTranspose(64, kernel_size=(2,2), strides=(2, 2), padding='same', name = 'Transpose_T04')(c19)
concat04 = layers.concatenate([t04, c1], axis=3, name= 'Concat_4')

c20 = layers.Conv2D(64, activation='relu', kernel_size=(3,3), name = 'C20', padding = 'same')(concat04)
Drop = layers.Dropout(0.1)(c20)
c21 = layers.Conv2D(64, activation='relu', kernel_size=(3,3), name = 'C21', padding = 'same')(Drop)

output = layers.Conv2D(1, activation='sigmoid', kernel_size=(1,1), name = 'OP')(c21)

UNet = tf.keras.Model(inputs=inputs, outputs=output, name='UNet')
```

```python
# Biên dịch mô hình U-Net
UNet.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
UNet.summary()
```

```python
early_stopping = EarlyStopping(
    monitor='val_loss',  # hoặc metric khác như 'val_accuracy'
    patience=10,
    restore_best_weights=True
)

# Huấn luyện U-Net
history_unet = UNet.fit(
    x_train, y_train,
    validation_data=(x_val, y_val),
    batch_size=20,
    epochs=100,
    verbose=1,
    callbacks=[early_stopping]
)

# Đánh giá U-Net
loss, accuracy = UNet.evaluate(x_test, y_test)
print(f"Loss: {loss}, Accuracy: {accuracy}")
```

**Xây dựng mô hình Efficientnet-B3**

```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras.applications.efficientnet import EfficientNetB3, preprocess_input

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import optimizers
import tensorflow.keras.callbacks as callbacks

# Hàm tải ảnh
def load_images(path):
    """
    Hàm để tải ảnh từ đường dẫn và resize về kích thước 256x256.
    """
    images = []
    for file_name in sorted(os.listdir(path)):
        img = cv2.imread(os.path.join(path, file_name), cv2.IMREAD_GRAYSCALE)
        if img is not None:
            img = cv2.resize(img, (256, 256))  # Resize về 256x256
            images.append(img)
    return np.array(images)

# Đường dẫn dữ liệu
tb_images_path = '/content/drive/MyDrive/Test_KLTN/TB_Chest_Radiography_Database/Tuberculosis'
control_images_path = '/content/drive/MyDrive/Test_KLTN/TB_Chest_Radiography_Database/Normal'

# Tải dữ liệu
tb_images = load_images(tb_images_path)
control_images = load_images(control_images_path)

# Gắn nhãn
x_data_class = np.concatenate((tb_images, control_images), axis=0)
y_data_class = np.concatenate((np.ones(len(tb_images)), np.zeros(len(control_images))), axis=0)

# Chia tập dữ liệu
x_train_class, x_temp_class, y_train_class, y_temp_class = train_test_split(
    x_data_class, y_data_class, test_size=0.2, stratify=y_data_class, random_state=42)

x_val_class, x_test_class, y_val_class, y_test_class = train_test_split(
    x_temp_class, y_temp_class, test_size=0.5, stratify=y_temp_class, random_state=42)

# Kiểm tra phân phối nhãn
print("Train class distribution:", np.bincount(y_train_class.astype(int)))
print("Validation class distribution:", np.bincount(y_val_class.astype(int)))
print("Test class distribution:", np.bincount(y_test_class.astype(int)))
```

**Phân đoạn Phổi trên dữ liệu phân loại**

```python
from tensorflow.keras.models import load_model

# Tải mô hình U-Net đã huấn luyện
unet_model = load_model("/content/drive/MyDrive/Test_KLTN/model_unet_v2.h5")

# Hàm phân đoạn vùng phổi
def segment_lungs(unet_model, images):
    images = images[..., np.newaxis] / 255.0  # Chuẩn hóa và thêm chiều
    predicted_masks = unet_model.predict(images)
    binary_masks = (predicted_masks > 0.5).astype(np.uint8)
    segmented_images = images * binary_masks  # Lấy vùng phổi
    return segmented_images.squeeze()  # Loại bỏ chiều không cần thiết

# Phân đoạn vùng phổi
x_train_ready = segment_lungs(unet_model, x_train_class)
x_val_ready = segment_lungs(unet_model, x_val_class)
x_test_ready = segment_lungs(unet_model, x_test_class)
```

```python
# Tiền xử lỹ dữ liệu cho mô hình phân loại
def preprocess_for_efficientnet(images, target_size=(300, 300)):
    """
    Tiền xử lý ảnh:
    1. Resize từ kích thước 256x256 về 300x300
    2. Chuyển ảnh từ grayscale sang RGB
    3. Chuẩn hóa theo chuẩn của EfficientNet
    """
    processed_images = []
    for img in images:
        img_resized = cv2.resize(img, target_size)  # Resize từ 256x256 về 300x300
        # Convert the image to 8-bit unsigned integer before color conversion
        img_resized = (img_resized * 255).astype(np.uint8)
        img_rgb = cv2.cvtColor(img_resized, cv2.COLOR_GRAY2RGB)  # Chuyển sang RGB
        img_normalized = preprocess_input(img_rgb)  # Chuẩn hóa
        processed_images.append(img_normalized)
    return np.array(processed_images)

# Áp dụng tiền xử lý
x_train_ready = preprocess_for_efficientnet(x_train_ready)
x_val_ready = preprocess_for_efficientnet(x_val_ready)
x_test_ready = preprocess_for_efficientnet(x_test_ready)

# Kiểm tra kích thước
print(f"x_train_ready shape: {x_train_ready.shape}")
print(f"x_val_ready shape: {x_val_ready.shape}")
print(f"x_test_ready shape: {x_test_ready.shape}")
```

**Xây dựng mô hình Efficientnet-B3**

```python
from tensorflow.keras.applications import EfficientNetB3
```

```python
# Xây dựng mô hình EfficientNet-B3
base_model = EfficientNetB3(include_top=False, input_shape=(300, 300, 3), pooling='avg', weights='imagenet')
x = Dropout(0.5)(base_model.output)  # Áp dụng Dropout
output = Dense(1, activation='sigmoid')(x)

efficientnet_model = Model(inputs=base_model.input, outputs=output)

# Compile mô hình
efficientnet_model.compile(
    optimizer=optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy', 'AUC', 'Precision', 'Recall']
)

# Callback
early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
reduce_lr = callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, min_lr=1e-6)
model_checkpoint = callbacks.ModelCheckpoint('best_efficientnet_model.keras', monitor='val_loss', save_best_only=True, mode='min')

# Tính toán class weight để cân bằng lớp
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train_class),
    y=y_train_class
)
class_weights = dict(enumerate(class_weights))

# Huấn luyện mô hình
history = efficientnet_model.fit(
    x_train_ready, y_train_class,
    validation_data=(x_val_ready, y_val_class),
    epochs=100,
    batch_size=32,
    callbacks=[early_stopping, reduce_lr, model_checkpoint],
    class_weight=class_weights
)
```

```python
efficientnet_model.summary()
```

```python
# Đánh giá mô hình trên tập kiểm tra
test_loss, test_accuracy, test_auc, test_precision, test_recall = efficientnet_model.evaluate(x_test_ready, y_test_class)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
print(f"Test AUC: {test_auc}")
print(f"Test Precision: {test_precision}")
print(f"Test Recall: {test_recall}")
print(f"Test F1 Score: {2 * (test_precision * test_recall) / (test_precision + test_recall)}")
```

```python
# Trực quan hóa lịch sử huấn luyện
# Vẽ biểu đồ độ chính xác
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title('Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Dự đoán trên tập test
y_pred_prob = efficientnet_model.predict(x_test_ready)  # Dự đoán xác suất
y_pred = (y_pred_prob > 0.5).astype(int)  # Chuyển thành nhãn 0 hoặc 1

# Tính ma trận nhầm lẫn
cm = confusion_matrix(y_test_class, y_pred)

# Vẽ ma trận nhầm lẫn
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Normal', 'Tuberculosis'])
disp.plot(cmap=plt.cm.Blues)

# Thêm tiêu đề và hiển thị
plt.title("Confusion Matrix - EfficientNet-B3")
plt.show()
```

```python
# Hàm Grad-CAM
def grad_cam(model, img_array, layer_name):
    grad_model = Model([model.inputs], [model.get_layer(layer_name).output, model.output])

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        loss = predictions[:, 0]  # Tính loss từ đầu ra (chỉ có 1 lớp)

    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    heatmap = tf.reduce_sum(tf.multiply(pooled_grads, conv_outputs[0]), axis=-1)

    heatmap = np.maximum(heatmap, 0)
    heatmap /= np.max(heatmap)  # Chuẩn hóa heatmap
    return heatmap, predictions

# Hàm vẽ Grad-CAM
def plot_gradcam(image, segmented_image, gradcam_result, classification_label, confidence, title="Grad-CAM"):
    """
    Hiển thị ảnh gốc, ảnh phân đoạn và bản đồ Grad-CAM.
    """
    fig, ax = plt.subplots(1, 3, figsize=(15, 5))

    # Hiển thị ảnh gốc
    ax[0].imshow(image, cmap='gray')
    ax[0].set_title(f"Original Image\nClassification: {classification_label} ({confidence:.2f})")
    ax[0].axis('off')

    # Hiển thị ảnh phân đoạn
    ax[1].imshow(segmented_image, cmap='gray')
    ax[1].set_title("Segmented Lungs")
    ax[1].axis('off')

    # Hiển thị Grad-CAM
    ax[2].imshow(segmented_image, cmap='gray')
    ax[2].imshow(cv2.resize(gradcam_result, (segmented_image.shape[1], segmented_image.shape[0])), cmap='jet', alpha=0.5)
    ax[2].set_title(title)
    ax[2].axis('off')

    plt.tight_layout()
    plt.show()

```

```python
# Chọn một ảnh từ tập kiểm tra
test_index = 3

test_image = x_test_ready[test_index:test_index+1]  # Lấy 1 ảnh đã tiền xử lý và phân đoạn.
original_image = x_test_class[test_index]  # Ảnh gốc chưa tiền xử lý

# Phân đoạn vùng phổi
segmented_image = x_test_ready[test_index]  # Vùng phổi đã phân đoạn

# Áp dụng Grad-CAM và lấy dự đoán phân loại
heatmap_1, predictions = grad_cam(efficientnet_model, test_image, layer_name='top_conv')

# Phân loại dựa trên đầu ra
confidence = predictions.numpy()[0, 0]
classification_label = "Tuberculosis" if confidence > 0.5 else "Normal"

# Hiển thị Grad-CAM với kết quả phân loại
plot_gradcam(original_image, segmented_image, heatmap_1, classification_label, confidence)
```
